```python
#gender distribution
plt.figure(figsize=(5,5)) # used to control width and height and measured in inches
ax = sns.countplot(data=df,x="Gender",palette="rainbow")
# ax.bar_label(ax.containers[0]) #Seaborn sometimes creates multiple groups (containers)
for container in ax.containers: # method - 2
    ax.bar_label(container)
plt.title("Gender Distribution")
plt.show() # for normal python script we need to write it
```

# from the above chart the no of females in the data is more than the number of males

```python
gb=df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
gb # agg to do some calculations in some groups 
```

```python
# Heatmap is for patterns, bar plot is for comparison too many numerical and one categorical then go for heatmap
sns.heatmap(gb,annot=True) # annot is used to show values in the matrix of graph
plt.title("parent's Education vs Student's Education")
plt.show()
```

#from the above chart we can conclude that  parents education will effect the childs education 

```python
# df["ParentMaritalStatus"].value_counts()
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
gb1
```

```python
sns.heatmap(gb1,annot=True)
plt.title("Parent's Maritial Status vs Studen's Education")
plt.plot()
```

# from the above chart we have no relation between parent's martial status and ;student's scores  

```python
# to detect outliers we use boxplot. outliers - out of range values
# outliers are typically defined using the Interquartile Range (IQR):
sns.boxplot(x=df["MathScore"])
plt.show()
```

# Distribution of Ethinic Group

```python
# using LOC to count  df.loc[row_condition, column_selection] loc is used when we need to select columns mainly
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()
l = ["group A","group B","group C","group D","group E"]
mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mlist,labels=l,autopct="%1.2f%%")
plt.title("Distribution of EthnicGroup")
plt.show()
```
